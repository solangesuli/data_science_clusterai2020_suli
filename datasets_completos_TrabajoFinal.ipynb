```python
# importamos las librerías necesarias para trabajar.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#Otros 
import warnings
warnings.filterwarnings('ignore')
```

```python
aeroparque_df = pd.read_csv("./data/87582_Aerop.csv",delimiter=',', parse_dates=['fechas'], na_values="-99.9") #datos desde 1960-2017      
aeroparque_df
```

```python
bsas_df = pd.read_csv("./data/87585_bsas.csv",delimiter=',', parse_dates=['fechas'], na_values="-99.9") #datos desde 1960-2017
bsas_df
```

```python
#cargo archivos de precipitacion (ya estan separados por anio, mes, dia, prcp, tmax,tmin)
aeroparque_df_precip = pd.read_csv("./data/87582_aerop_precip.csv",delimiter=',', na_values="-99.9") #datos desde 1960-2017   
precip_aeroparque_df= aeroparque_df_precip[['year','month','day','prcp']] #archivos 1977-2018

bsas_df_precip= pd.read_csv("./data/87585_bsas_precip.csv",delimiter=',', na_values="-99.9") #datos desde 1960-2017   
precip_bsas_df=bsas_df_precip[['year','month','day','prcp']]
```

```python
#selecciono tmax y tmin del 2018 de los archivos de precipitacion
aerop_tmaxtmin = aeroparque_df_precip[['year','month','day','tmax','tmin']]
aerop_tmaxtmin_18 =aerop_tmaxtmin.loc[ (aerop_tmaxtmin['year']==2018)]

print(aerop_tmaxtmin_18)

bsas_tmaxtmin = bsas_df_precip[['year','month','day','tmax','tmin']]
bsas_tmaxtmin_18 =bsas_tmaxtmin.loc[ (bsas_tmaxtmin['year']==2018)]

print(bsas_tmaxtmin_18)
```

```python
#renombro las columnas con los mismos nombres que los otros df

#para los archivos de precipitacion
precip_aeroparque_df = precip_aeroparque_df.rename(columns={
    'year':'anio',
    'month':'mes',
    'day':'dia',
    'prcp':'precip'
})

precip_bsas_df = precip_bsas_df.rename(columns={
    'year':'anio',
    'month':'mes',
    'day':'dia',
    'prcp':'precip'
})

print(f'Head pp_aeroparque: \n{precip_aeroparque_df.head()}\r\n')
print(f'Head pp_bsas: \n{precip_bsas_df.head()}\r\n')


#para los archivos de tmax y tmin del 2018

aerop_tmaxtmin_18 = aerop_tmaxtmin_18.rename(columns={
    'year':'anio',
    'month':'mes',
    'day':'dia'
})

bsas_tmaxtmin_18 = bsas_tmaxtmin_18.rename(columns={
    'year':'anio',
    'month':'mes',
    'day':'dia'
})

print(f'Head tmaxtmin18_aeroparque: \n{aerop_tmaxtmin_18.head()}\r\n')
print(f'Head tmaxtmin18_bsas: \n{bsas_tmaxtmin_18.head()}\r\n')
```

```python
#separamos la columna fecha en: "anio", "mes" y "dia"
aeroparque_df['anio'] = aeroparque_df.fechas.dt.year
aeroparque_df['mes'] = aeroparque_df.fechas.dt.month
aeroparque_df['dia'] = aeroparque_df.fechas.dt.day

#reordeno las columnas del df y saco la columna de fechas (reesctibo el df)
aeroparque_df=aeroparque_df[['anio','mes','dia','precip','tmax','tmin']]
aeroparque_df
```

```python
#separamos la columna fecha en: "anio", "mes" y "dia"
bsas_df['anio'] = bsas_df.fechas.dt.year
bsas_df['mes'] = bsas_df.fechas.dt.month
bsas_df['dia'] = bsas_df.fechas.dt.day

#reordeno las columnas del df y saco la columna de fechas (reesctibo el df)
bsas_df=bsas_df[['anio','mes','dia','precip','tmax','tmin']]
bsas_df
```

```python
# Verificar existencia de Nulls
print(f'Any nulls in columns (Aeroparque): \n{aeroparque_df.isnull().any()}\r\n')
print(f'Any nulls in columns(Bs As): \n{bsas_df.isnull().any()}\r\n')
```

```python
#seleccionamos el periodo comun de los archivos de precipitacion y de temperatura (1977-2018)
aeroparque_df = aeroparque_df.loc[ (aeroparque_df['anio'] >= 1977) & (aeroparque_df['anio'] <= 2018) ]
bsas_df = bsas_df.loc[(bsas_df['anio'] >= 1977) & (bsas_df['anio'] <= 2018)]
```

```python
#Los nuevos df quedan asi: (observar que tienen distinta cant de filas!)
print(f'Info de Aeroparque:')
filas = np.shape(aeroparque_df)[0]
print(f'Filas: {filas}')
columnas = np.shape(aeroparque_df)[1]
print(f'Columnas: {columnas}\r\n')

print(f'Info de Bs As:')
filas = np.shape(bsas_df)[0]
print(f'Filas: {filas}')
columnas = np.shape(bsas_df)[1]
print(f'Columnas: {columnas}\r\n')
```

```python
#reseteo los indices 
aeroparque_df = aeroparque_df.reset_index(drop=True)
aeroparque_df
```

```python
#reseteo los indices
bsas_df = bsas_df.reset_index(drop=True)
bsas_df
```

```python
#genero matriz unicamente con fechas para mergear (1977-2018) -> periodo de 42 anios (deberia haber 15340 datos (filas))
fechas_df = pd.date_range(start='1977-01-01', end='2018-12-31', freq='D')
fechas_df = pd.DataFrame(fechas_df)
fechas_df.columns = ['fechas']

#separo las columnas igual que en los otros df
fechas_df['anio'] = fechas_df.fechas.dt.year
fechas_df['mes'] = fechas_df.fechas.dt.month
fechas_df['dia'] = fechas_df.fechas.dt.day
fechas_df = fechas_df [['anio','mes','dia']]

fechas_df
```

```python
#mergeo fechas_df con aeroparque_df

aeroparque_df_completo = pd.merge(aeroparque_df, fechas_df, how='outer', on=['anio','mes','dia'])
aeroparque_df_completo #notar que tiene 15340 filas y en los ultimos anios que no teniamos datos puso NaNs
```

```python
#mergeo fechas_df con bsas_df
bsas_df_completo = pd.merge(bsas_df, fechas_df, how='outer', on=['anio','mes','dia'])
bsas_df_completo #notar que tiene 15340 filas y en los ultimos anios que no teniamos datos puso NaNs
```

```python
#reemplazo la columna de precipitacion en ambos df
aeroparque_df_completo = pd.merge(aeroparque_df_completo, precip_aeroparque_df, how='outer', on=['anio','mes','dia']) 
aeroparque_df_completo = aeroparque_df_completo[['anio','mes','dia','precip_y','tmax','tmin']]
aeroparque_df_completo.rename(columns={'precip_y':'precip'},inplace=True)
print(aeroparque_df_completo)


bsas_df_completo = pd.merge(bsas_df_completo, precip_bsas_df, how='outer', on=['anio','mes','dia']) 
bsas_df_completo = bsas_df_completo[['anio','mes','dia','precip_y','tmax','tmin']]
bsas_df_completo.rename(columns={'precip_y':'precip'},inplace=True)
print(bsas_df_completo)
```

```python
#adherimos las temepraturas del 2018 al df completo
aero_temps = aeroparque_df_completo.drop('precip', axis=1)
df = aeroparque_df_completo
df.loc[df.anio > 2017, ['tmin', 'tmax']] = aerop_tmaxtmin_18[['tmin', 'tmax']]
aeroparque_df_completo = df.copy()
aeroparque_df_completo
```

```python
#agregamos columna dia del anio
fechas_df = pd.date_range(start='1977-01-01', end='2018-12-31', freq='D')
aeroparque_df_completo['diaDelAnio'] = fechas_df.dayofyear
```

```python
#adherimos las temepraturas del 2018 al df completo
bsas_temps = bsas_df_completo.drop('precip', axis=1)
df2 = bsas_df_completo
df2.loc[df.anio > 2017, ['tmin', 'tmax']] = aerop_tmaxtmin_18[['tmin', 'tmax']]
bsas_df_completo = df2.copy()
bsas_df_completo
```

```python
#agregamos columna dia del anio
fechas_df = pd.date_range(start='1977-01-01', end='2018-12-31', freq='D')
bsas_df_completo['diaDelAnio'] = fechas_df.dayofyear
```
